```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 15, 5
import numpy as np

labels = ["LU для обычной матрицы", "LU для симметрической матрицы", "LDL^T для симметрической матрицы"]
idx = 0
with open('times.txt') as f:
    for row in f.readlines():
        plt.plot([i for i in range(10, 2010, 10)], [i / 1000 for i in list(map(float, row.split()))], label=labels[idx])
        plt.legend()
        idx += 1
    
plt.show()
```
